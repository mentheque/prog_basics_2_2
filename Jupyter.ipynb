## Задача:

```python
import numpy as np

def task_chukh(n):
    vert = np.array([i for i in range(n)])*np.ones((n, n), dtype = int)
    ret = np.array([i%2 + 1 for i in range(n)])*np.ones((n, n), dtype = int)
    ret[vert >= vert.transpose()] = ret.transpose()[vert >= vert.transpose()]
    return ret

print(task_chukh(10))
```

```python
import requests
from IPython.display import display, Image

def print_comics(num):
    com = requests.get(f"https://xkcd.com/{num}/info.0.json").json()
    print(com["title"])
    print(f"{com['day']}.{com['month']}.{com['year']}")
    display(Image(com["img"]))
    print(com["alt"])
    
print("Demonstration:")
print("___614___")
print_comics(614)
print("----------")

print("___641___")
print_comics(641)
print("----------")

print("___416___")
print_comics(416)
```
